# Asymmetric Cryptography

One way to authenticate is to prove that you know a &ldquo;secret&rdquo;. In this context, you can prove that the same person who proved knowledge of a secret in the past is proving knowledge of the secret once more. 

Asymmetric cryptography makes it possible for someone to prove possession of a secret without of revealing it. 


## 1. The Foundation for Asymmetric Cryptography 

Suppose that the function $f(\cdot)$ takes values from a set $\mathcal{S}$ and sends them to a set $\mathcal{P}$ that might be different from $\mathcal{S}$. 

For example, given a prime number $q$, the set $\mathcal{S}$ might be the set of $q$ integers $$s \in \{0, 1, 2, ... q-1\}.$$

For a different prime number $r$, let the set $\mathcal{P}$ be the set of pairs $(P_x, P_y)$ where each individual value satisfies $$P_x, P_y \in \{0, 1, 2, ... r-1\}$$

In this type of mathematical notation, the face of the font matters. An uppercase P in the caligraphic font, $\mathcal{P}$ is a set. An upppercase $P$ in an italic font is a pair in the set: $$P \in \mathcal{P}$$

In this notation, a secret key will a number $s \in \mathcal{S}$. A public key will be a pair $(P_x, P_y) \in \mathcal{P}$. 

### 1.1 Addition in $\mathcal{S}$

For $s, t \in \mathcal{S},$ we can define define a value $s + t$ that is also in $\mathcal{S}$ using modular arithmetic. 

What this means is take the usual sum of $s$ and $t$ as ordinary integers. This sum might not be in $\mathcal{S}$. To get a value that is in $\mathcal{S}$, the take the residual with respect to the modulus $q$. To calculate the sum in Python, use the expression

`(s + t) % q` 

The parentheses in this expression help you, the reader, know what order to use in evaluating this expression. You add the two numbers first, then take the residual.

### 1.3 Defining Addition in $\mathcal{P}$ 

There is a natural way to define addition in $\mathcal{P}.$ This is not the definition we want to use. Instead, we want to pick a very special form of addition that is defined only for a subset of all the points in $\mathcal{P}.$ This will help us define a function $f:\mathcal{S} \rightarrow \mathcal{P}$ that satisfies the two properties that we'll need to create a crypto-system. Let&rsquo;s restate those properties. 


### 1.4 Properties for $f(\cdot)$

For a function $f:\mathcal{S} \rightarrow \mathcal{P}:$ 

(i) We will say that $f$ &ldquo;preserves addition&rdquo; if for any $s, t \in \mathcal{S}$, $$f(s +t) = f(s) + f(t).$$ 

(ii) We will say that $f$ is a &ldquo;one-way function&rdquo; if the fastest way to solve an equation of the form $$P = f(s)$$ is the &ldquo;brute force &rdquo; approach that calculates $f(s)$ for a sequence of values in $\mathcal{S}$ and stops when it finds one that satisfies the equation. 

It is easy to come up with functions that satisfy one of these two properties. 

- Linear functions preserve addition.

- Hash functions are one way functions. 

It is much more challenging to come up with functions that satisfy both properties. The extra flexibility that we can exploit to do this comes from the fact that we have not yet defined what it means to add two pairs of integers from $\mathcal{P}.$



### 1.5 Defining Addition in $\mathcal{P}$

The presentation here will not describe how mathematicians discovered a definition of addition in $\mathcal{P}$ that makes it possible to find a function $f$ that satisfies the two required properties. There are many sources that provide this information. 

Here we will simply state the definition. It starts with a specific example from a family of curves known as &ldquo;elliptic curves&rdquo;. One such curve can be characterized using a function $\mathcal{E}.$ To be precise about the definition of that function, it helps to have a way to refer to the set of non-negative integers less than some prime number $r$. One widely used symbol is $\mathbb{F}_r$. This uses a version of the capital letter F because such sets are called &ldquo;finite fields.&rdquo; For now, don&rsquo;t worry about what a field is. We'll explain later.   

The function $\mathcal{E}$ takes elements in $\mathcal{P}$ as inputs and it returns elements in $\mathbb{F}_r$. An element in $\mathcal{P}$ is just a pair of elements in $\mathbb{F}_r$. This allows the following simple definition of the function $\mathcal{E}.$ 


#### The Curve
Let $r$ be the prime number $2^{255} - 19$. Define the function $\mathcal{E}: \mathbb{F}_r \times \mathbb{F}_r \rightarrow \mathbb{F}_r$ by the rule 

$$\mathcal{E}(x,y) = y^2 - x^2 - d x^2 y^2 -1 \pmod r$$
where 
$$d = - \frac{121665}{121666} \pmod r$$

Define the &ldquo;curve&rdquo; $\mathscr{C}$ as the set of points in $\mathcal{P}$ that satisfy the equation $$\mathcal{E}(P_x, P_y) = 0.$$


It is misleading and confusing to call a set that contains a bunch of pairs of integers a &ldquo;curve&rdquo;. Unfortunately, this is the accepted terminology. It is a little bit easier to understand if you think of $\mathcal{E}$ as a function that is defined on pairs of real numbers, we could draw a picture of the the set of points that satisfy this equation. This picture would look like a smooth curve. 

But explanations that cheat by treating $x$ and $y$ as if they were real numbers are also misleading. It is better to stay focused on the fact that all the symbols in this equation have to be integers. 

Then expression that should jump out at you is the definition of $d.$ How can this ratio be an integer? And how can we calculate it? 

#### Inverses of Integers

When we work in the set $\mathbb{F}_r$ we use modular arithmetic to calculate sums and products. It is simplest if we stick with Python to specify how how the basic operations are defined. As noted above, the sum of two elemens $x, y \in \mathbb{F}_p$ can be calculated as 

`(x + y) %r`

By analogy, the product of two numbers is 

`(x*y) % r` 

For any element $x \in \mathbb{F}_r$ that is not equal to zero, the definition of the expression $$z = \frac{1}{x} \in \mathbb{F}_r$$ follows immediately. It is the element in $\mathbb{F}_r$ that satisifes  $$z * x  =  \frac{1}{x} * x = 1 \pmod r$$ 

In Python, the element $\frac{1}{x}$ is the positive integer $z$ that is strictly less than $r$ that satisfies 

`(z * x) % r == 1`

#### Finite Fields

How do we know that there will be such an element? 

First, let's define the concept of a &ldquo;field&rdquo; that we mentioned above. A &ldquo;field&rdquo; is a set where for any nonzero element $x,$ there exists an inverse $\frac{1}{x}$ such that $$\frac{1}{x} * x = 1.$$

To rephrase the question, how do we know that there are any fields that are made up of a finite set of integers? 

A famous theorem in mathematics shows that $\mathbb{F}_r$ is a finite field if $r$ is a prime number. A related result known as &ldquo;Fermat&rsquo;s Little Theorem&rdquo; tells us how to calculate the inverse: 

THEOREM: If $r$ is prime and $x \neq 0$ is an element of $\mathbb{F}_r,$ $$\frac{1}{x} = x^{r-2}$$

## 4. Cleaning Up the Notation

```python
def bin_rep(m):
    z = m
    bin_bit = {}
    for j in range(255, -1, -1): 
        bin_bit[j], z  = divmod(z, 2**j)
    return bin_bit

def value_from(br):
    s = 0
    for k, v in br.items():
        s += 2**k * v 
    return s 


class CurvePoint2:
    def __init__(self, x, y):
        self.x = x % r
        self.y = y % r

    def __eq__(self, N):
        if all([
                isinstance(N, CurvePoint), 
                self.x == N.x,
                self.y
            ]):
            return True 
        else: 
            return False

    def two_to_power_j_times_(self): 
        mult_of_ = {}
        mult_of_[0] = self
        for j in range(1, 255):
            mult_of_[j] = cp_add( mult_of_[j-1], mult_of_[j-1] )
        return mult_of_
    
    def scalar_mult_(self, m):
        m = m % r
        dict = self.two_to_power_j_times_()
        cp = CurvePoint(0,1)
        for j in range(255):
            if bin_rep(m)[j]: 
                cp = cp_add(dict[j], cp)
        return cp
        
    def __rmatmul__(self, m):
        return self.scalar_mult_(m)

    
```
